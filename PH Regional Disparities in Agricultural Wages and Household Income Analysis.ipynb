## Feature Engineering

### Creating farm_income_share

```python
df['farm_income_share'] = df['avg_annual_farm_incm_farm_households_02_03']/df['avg_annual_total_incm_farm_households_02_03']
df['farm_income_share'].head(16)
```

### Creating off-farm_income_share

```python
df['off_farm_income_share'] = df['avg_annual_off_farm_incm_farm_households_02_03']/df['avg_annual_total_incm_farm_households_02_03']
df['off_farm_income_share'].head(16)
```

### Creating non-farm_income_share

```python
df['non_farm_income_share'] = df['avg_annual_non_farm_incm_farm_households_02_03']/df['avg_annual_total_incm_farm_households_02_03']
df['non_farm_income_share'].head(16)
```

### Creating other_sources_income_share

```python
df['other_sources_income_share'] = df['avg_annual_other_sources_incm_farm_households_02_03']/df['avg_annual_total_incm_farm_households_02_03']
df['other_sources_income_share'].head(16)
```
